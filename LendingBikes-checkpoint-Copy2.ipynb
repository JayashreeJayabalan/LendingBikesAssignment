```python
# List of variables to map

varlist =  ['season']

# Defining the map function
def season_map(x):
    return x.map({'1': 'spring', "2": 'summer', "3" : 'fall', "4" : 'winter'})

# Applying the function to the housing list
df[varlist] = df[varlist].apply(season_map)
```

```python
df.head()
```

```python
df['weathersit'] = df['weathersit'].map(str)
```

```python
# List of variables to map
#Want to add type of weathersit to fields for better readability
varlist =  ['weathersit']

# Defining the map function
def weather_map(x):
    return x.map({'1': 'clearsky', "2": 'Mist', "3" : 'lightsnow', "4" : 'heavyrain'})

# Applying the function to the housing list
df[varlist] = df[varlist].apply(weather_map)
```

```python
df.weekday.unique()
```

```python
df['weekday'] = df['weekday'].map(str)
```

```python
# List of variables to map
#Want to add names of weekday to fields for better readability
varlist =  ['weekday']

# Defining the map function
def weekday_map(x):
    return x.map({'0': 'TUE', "1": 'WED', "2" : 'THU', "3" : 'FRI', '4' : 'SAT', '5': 'SUN','6' :'MON' })

# Applying the function to the housing list
df[varlist] = df[varlist].apply(weekday_map)
```

# Reading, Understanding and Cleaning Data 

1.We do not need instant as it is just an ID.

2.dteday field has already been extracted into Year, month. The weekday, working day and holiday are further giving information   about the day. Hence we dont need the date field.

3.Also casual and registered are adding to cnt. As cnt is the target variable, it is better to remove casual and registered.

```python
df.drop(['instant', 'dteday', 'casual', 'registered'],  axis=1, inplace=True)
df.head()
```

### Check variance to see the correlation of variables before dropping

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
